# **Data Reperesentation**

```python
import pandas as pd
import matplotlib.pyplot as plt
data = ['Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic',
        'Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic',
        'Coke Classic','Coke Classic','Coke Classic','Coke Classic','Coke Classic','Diet Coke','Diet Coke','Diet Coke',
        'Diet Coke','Diet Coke','Diet Coke','Diet Coke','Diet Coke','Dr.Pepper','Dr.Pepper','Dr.Pepper','Dr.Pepper',
        'Dr.Pepper','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi','Pepsi',
        'Pepsi','Sprite','Sprite','Sprite','Sprite','Sprite']
df = pd.Series(data).value_counts()
df
```

```python
d = pd.DataFrame(df)
d
```

```python
d['rf'] = d / len(data)
d['prf'] = d['rf'] * 100
d['cf'] = df.cumsum()
d['rcf'] = d['cf'] / df.sum()
d['prcf'] = d['rcf'] * 100
d
```

```python
freq = d[0].plot(kind='bar', title='Bar Graph', legend=True, fontsize=12)
freq.set_xlabel('Soft Drinks', fontsize=12)
freq.set_ylabel('Frequency', fontsize=12)
plt.show()
```

```python
RelFreq = d['rf'].plot(kind='bar', title='Bar Graph', legend=True, fontsize=12)
plt.xlabel('Soft Drinks', fontsize=12)
plt.ylabel('Relative Frequency', fontsize=12)
plt.show()
```

```python
PercFreq = d['prf'].plot.pie(legend=True, figsize=(10,5),autopct="%.0f%%")
PercFreq.set_ylabel('Percent Relative Frequency')
plt.show()
```

```python
cf = d['cf'].plot.line()
plt.xlabel('Soft Drinks', fontsize=12)
plt.ylabel('Cumulative Frequency', fontsize=12)
plt.show()
```

```python
drink = pd.Series(data).unique()
x = drink
y = d['rcf']
plt.xlabel('Soft Drinks')
plt.ylabel('Relative Cumulative Frequency')
plt.scatter(x, y)
```

```python
plt.style.use('ggplot')
plt.hist(data,bins=5)
plt.xlabel('Soft Drinks', fontsize=12)
plt.ylabel('Percent Relative Cumulative Frequency', fontsize=12)
plt.show()
```
