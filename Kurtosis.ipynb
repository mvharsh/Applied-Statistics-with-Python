# **Kurtosis**

**Kurtosis** is a measure of whether or not a distribution is heavy-tailed or light-tailed relative to a normal distribution.

image.png

The kurtosis of a **normal** distribution is **mesokurtic**.

If a given distribution has a kurtosis less than normal, it is said to be **playkurtic**, which means it tends to produce fewer and less extreme outliers than the normal distribution.

If a given distribution has a kurtosis greater than normal, it is said to be **leptokurtic**, which means it tends to produce more outliers than the normal distribution.
